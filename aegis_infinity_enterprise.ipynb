```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  CELL 1 — CORE CALIBRATION ENGINE                                   ║
# ╚══════════════════════════════════════════════════════════════════════╝

def _logit(c):
    c = np.clip(c, 1e-7, 1 - 1e-7)
    return np.log(c / (1 - c))

# ── Metrics ────────────────────────────────────────────────────────────

def ece(confs, labels, n_bins=15):
    """Expected Calibration Error"""
    bins = np.linspace(0, 1, n_bins + 1)
    err = 0.0
    for i in range(n_bins):
        m = (confs >= bins[i]) & (confs < bins[i + 1])
        if m.sum() < 2: continue
        err += (m.sum() / len(confs)) * abs(labels[m].mean() - confs[m].mean())
    return float(err)

def mce(confs, labels, n_bins=15):
    """Maximum Calibration Error"""
    bins = np.linspace(0, 1, n_bins + 1)
    errs = []
    for i in range(n_bins):
        m = (confs >= bins[i]) & (confs < bins[i + 1])
        if m.sum() < 2: continue
        errs.append(abs(labels[m].mean() - confs[m].mean()))
    return float(max(errs)) if errs else 0.0

def overconfidence_rate(confs, labels, threshold=0.8):
    """% of predictions where conf > threshold but wrong"""
    mask = confs > threshold
    if mask.sum() == 0: return 0.0
    return float((labels[mask] == 0).mean())

def calibration_report(confs, labels, model_name='Model'):
    """Full calibration report dict"""
    c = np.asarray(confs, float)
    l = np.asarray(labels, float)
    return {
        'model':          model_name,
        'n_samples':      len(c),
        'ECE':            round(ece(c, l), 5),
        'MCE':            round(mce(c, l), 5),
        'Brier':          round(brier_score_loss(l, c), 5),
        'LogLoss':        round(log_loss(l, c), 5),
        'AUC-ROC':        round(roc_auc_score(l, c), 5),
        'Overconf@80%':   round(overconfidence_rate(c, l, 0.8), 5),
        'MeanConf':       round(c.mean(), 4),
        'MeanAcc':        round(l.mean(), 4),
        'ConfGap':        round(c.mean() - l.mean(), 5),  # + = overconfident
    }

# ── Calibration Methods ────────────────────────────────────────────────

class CalibrationMethod:
    def fit(self, conf_cal, labels_cal): ...
    def transform(self, conf_test): ...

class TemperatureScaling(CalibrationMethod):
    name = 'Temperature Scaling'
    def fit(self, cc, lc):
        cc, lc = np.asarray(cc, float), np.asarray(lc, float)
        res = minimize_scalar(
            lambda T: ece(sigmoid(_logit(cc) / T), lc),
            bounds=(0.05, 20.0), method='bounded')
        self.T_ = res.x
        return self
    def transform(self, ct):
        return np.clip(sigmoid(_logit(np.asarray(ct, float)) / self.T_), 1e-7, 1-1e-7)
    def params(self): return {'T': round(self.T_, 4)}

class IsotonicScaling(CalibrationMethod):
    name = 'Isotonic Regression'
    def fit(self, cc, lc):
        self.iso_ = IsotonicRegression(out_of_bounds='clip')
        self.iso_.fit(np.asarray(cc, float), np.asarray(lc, float))
        return self
    def transform(self, ct):
        return np.clip(self.iso_.predict(np.asarray(ct, float)), 1e-7, 1-1e-7)
    def params(self): return {'knots': len(self.iso_.X_thresholds_)}

class BetaScaling(CalibrationMethod):
    name = 'Beta Calibration'
    def fit(self, cc, lc):
        cc, lc = np.asarray(cc, float), np.asarray(lc, float)
        lp = np.log(np.clip(cc, 1e-7, 1-1e-7))
        lm = np.log(1 - np.clip(cc, 1e-7, 1-1e-7))
        def nll(p):
            prob = np.clip(sigmoid(p[0]*lp + p[1]*lm + p[2]), 1e-7, 1-1e-7)
            return -np.mean(lc * np.log(prob) + (1-lc) * np.log(1-prob))
        res = minimize(nll, [1., -1., 0.], method='Nelder-Mead',
                       options={'xatol':1e-6,'fatol':1e-6,'maxiter':2000})
        self.a_, self.b_, self.c_ = res.x
        return self
    def transform(self, ct):
        ct = np.asarray(ct, float)
        lp = np.log(np.clip(ct, 1e-7, 1-1e-7))
        lm = np.log(1 - np.clip(ct, 1e-7, 1-1e-7))
        return np.clip(sigmoid(self.a_*lp + self.b_*lm + self.c_), 1e-7, 1-1e-7)
    def params(self): return {'a': round(self.a_,4), 'b': round(self.b_,4), 'c': round(self.c_,4)}

METHODS = {
    'beta':        BetaScaling,
    'isotonic':    IsotonicScaling,
    'temperature': TemperatureScaling,
}

# ── TruthGuard Core ────────────────────────────────────────────────────

DOMAIN_THRESHOLDS = {
    'general':   0.60,
    'financial': 0.78,
    'legal':     0.85,
    'medical':   0.92,
    'nuclear':   0.98,
    'custom':    None,   # user-defined
}

class TruthGuard:
    """
    AEGIS-∞ TruthGuard Engine
    Calibrates model confidence and makes ANSWER / ABSTAIN decisions.
    Supports: Beta · Isotonic · Temperature scaling
    Domains:  general · financial · legal · medical · nuclear · custom
    """

    def __init__(self, method='beta', domain='medical', custom_threshold=None):
        self.method    = method
        self.domain    = domain
        self.threshold = custom_threshold if domain == 'custom' else DOMAIN_THRESHOLDS[domain]
        self._cal      = METHODS[method]()
        self._fitted   = False
        self._history  = []   # audit trail

    def fit(self, conf_cal, labels_cal):
        self._cc = np.asarray(conf_cal, float)
        self._lc = np.asarray(labels_cal, float)
        self._cal.fit(self._cc, self._lc)
        self._fitted = True
        self._pre_report  = calibration_report(self._cc, self._lc, 'Pre-Cal')
        cal_cc = self._cal.transform(self._cc)
        self._post_report = calibration_report(cal_cc, self._lc, 'Post-Cal')
        return self

    def predict(self, conf_raw, context=''):
        assert self._fitted, 'Call .fit() first'
        c_raw = np.atleast_1d(np.asarray(conf_raw, float))
        c_cal = self._cal.transform(c_raw)
        results = []
        for raw, cal in zip(c_raw, c_cal):
            abstain = bool(cal < self.threshold)
            margin  = float(cal - self.threshold)
            r = {
                'conf_raw':        round(float(raw), 4),
                'conf_calibrated': round(float(cal), 4),
                'threshold':       round(self.threshold, 4),
                'verdict':         'ABSTAIN' if abstain else 'ANSWER',
                'risk_level':      self._risk(cal),
                'margin':          round(margin, 4),
                'domain':          self.domain,
                'method':          self.method,
                'context':         context,
                'timestamp':       time.strftime('%Y-%m-%d %H:%M:%S'),
                'audit_id':        hashlib.md5(f"{raw}{time.time()}".encode()).hexdigest()[:8],
            }
            self._history.append(r)
            results.append(r)
        return results[0] if len(results) == 1 else results

    def _risk(self, cal):
        t = self.threshold
        if cal >= t + 0.05:          return 'LOW'
        elif cal >= t:               return 'MODERATE'
        elif cal >= t - 0.10:        return 'HIGH'
        else:                        return 'CRITICAL'

    def reliability_curve(self, n_bins=10):
        bins = np.linspace(0, 1, n_bins + 1)
        rows = []
        for i in range(n_bins):
            m = (self._cc >= bins[i]) & (self._cc < bins[i+1])
            if m.sum() < 2: continue
            rows.append({
                'bin_center': (bins[i]+bins[i+1])/2,
                'accuracy':   self._lc[m].mean(),
                'confidence': self._cc[m].mean(),
                'count':      int(m.sum()),
            })
        return pd.DataFrame(rows)

    def audit_trail(self):
        return pd.DataFrame(self._history)

    def summary(self):
        return {
            'pre_calibration':  self._pre_report,
            'post_calibration': self._post_report,
            'improvement':      {
                'ECE_delta':    round(self._pre_report['ECE'] - self._post_report['ECE'], 5),
                'Brier_delta':  round(self._pre_report['Brier'] - self._post_report['Brier'], 5),
            },
            'method_params': self._cal.params(),
        }

banner('✓ Calibration Engine Loaded',
       'ECE · MCE · Brier · AUC-ROC · Beta · Isotonic · Temperature')
```

```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  CELL 2 — SYNTHETIC DATA GENERATOR (Enterprise Scenarios)           ║
# ╚══════════════════════════════════════════════════════════════════════╝
# Simulates real AI model outputs across different deployment scenarios.
# Replace with your own model's (confidence, label) pairs.

class ScenarioFactory:
    """
    Generates realistic model confidence distributions.
    Scenarios model common failure modes seen in production AI.
    """

    SCENARIOS = {
        'well_calibrated': {
            'desc': 'Model is nearly perfectly calibrated (ideal case)',
            'fn': lambda n, rng: (
                rng.beta(6, 2, n),
                lambda c: rng.uniform(0,1,n) < c - rng.normal(0, 0.03, n)
            )
        },
        'overconfident': {
            'desc': 'Model consistently overestimates its confidence (common in LLMs)',
            'fn': lambda n, rng: (
                np.clip(rng.beta(8, 1.5, n), 0.55, 0.99),
                lambda c: rng.uniform(0,1,n) < c - 0.25 + rng.normal(0, 0.05, n)
            )
        },
        'underconfident': {
            'desc': 'Model underestimates itself — often after RLHF over-tuning',
            'fn': lambda n, rng: (
                np.clip(rng.beta(2, 5, n), 0.01, 0.65),
                lambda c: rng.uniform(0,1,n) < c + 0.20 + rng.normal(0, 0.04, n)
            )
        },
        'bimodal': {
            'desc': 'Model is certain on easy cases, confused on hard ones',
            'fn': lambda n, rng: (
                np.concatenate([
                    rng.beta(9, 1, n//2),
                    rng.beta(2, 8, n - n//2)
                ]),
                lambda c: rng.uniform(0,1,n) < np.clip(c, 0.05, 0.95)
            )
        },
        'domain_shift': {
            'desc': 'Model trained on one domain, deployed in another',
            'fn': lambda n, rng: (
                rng.beta(7, 1.5, n),
                lambda c: rng.uniform(0,1,n) < c - 0.35 + rng.normal(0, 0.08, n)
            )
        },
    }

    @classmethod
    def generate(cls, scenario='overconfident', n=500, seed=2026):
        rng = np.random.RandomState(seed)
        cfg = cls.SCENARIOS[scenario]
        confs, label_fn = cfg['fn'](n, rng)
        confs = np.clip(confs, 0.01, 0.99)
        labels = np.clip(label_fn(confs).astype(int), 0, 1)
        return confs, labels, cfg['desc']

    @classmethod
    def all_scenarios(cls, n=400, seed=2026):
        results = {}
        for name in cls.SCENARIOS:
            c, l, desc = cls.generate(name, n, seed)
            results[name] = {'confs': c, 'labels': l, 'desc': desc}
        return results

# Demonstrate all scenarios
all_data = ScenarioFactory.all_scenarios(n=400)

_html('''
<div style="background:#0d1321;border:1px solid #1e3a5f;border-radius:8px;
            padding:16px 20px;font-family:monospace;">
  <div style="color:#00e5ff;font-weight:bold;margin-bottom:10px;">⬡ SCENARIO LIBRARY LOADED</div>
  <table style="width:100%;border-collapse:collapse;font-size:11px;">
    <tr style="color:#ffd700;border-bottom:1px solid #1e3a5f;">
      <td style="padding:4px 8px;">Scenario</td>
      <td style="padding:4px 8px;">Description</td>
      <td style="padding:4px 8px;">N</td>
      <td style="padding:4px 8px;">Mean Conf</td>
      <td style="padding:4px 8px;">Mean Acc</td>
    </tr>''' +
    ''.join(f'''
    <tr style="border-bottom:1px solid #1e3a5f22;">
      <td style="padding:4px 8px;color:#00e5ff;">{name}</td>
      <td style="padding:4px 8px;color:#e8f4f8;">{d["desc"]}</td>
      <td style="padding:4px 8px;color:#4a6280;">{len(d["confs"])}</td>
      <td style="padding:4px 8px;color:#ffd700;">{d["confs"].mean():.3f}</td>
      <td style="padding:4px 8px;color:#00ff88;">{d["labels"].mean():.3f}</td>
    </tr>''' for name, d in all_data.items()) +
  '</table></div>')
```

```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  CELL 3 — VISUALIZATION ENGINE                                      ║
# ╚══════════════════════════════════════════════════════════════════════╝

def plot_reliability_diagram(tg, title='Reliability Diagram', ax_main=None, ax_hist=None):
    """Publication-quality reliability diagram with confidence histogram"""
    standalone = ax_main is None
    if standalone:
        fig = plt.figure(figsize=(8, 6), facecolor=P['bg'])
        gs = gridspec.GridSpec(2, 1, height_ratios=[3,1], hspace=0.06)
        ax_main = fig.add_subplot(gs[0])
        ax_hist = fig.add_subplot(gs[1])

    df = tg.reliability_curve(15)
    cal_confs = tg._cal.transform(tg._cc)

    # Perfect calibration
    ax_main.plot([0,1],[0,1], '--', color=P['gold'], lw=1.5,
                 alpha=0.6, label='Perfect calibration', zorder=1)
    ax_main.fill_between([0,1],[0,1],[0,1], alpha=0.03, color=P['gold'])

    # Pre-calibration
    ax_main.plot(df['confidence'], df['accuracy'], 'o-',
                 color=P['red'], lw=2, ms=6, label='Before calibration', zorder=3)

    # Post-calibration
    frac_pos, mean_pred = calibration_curve(tg._lc, cal_confs, n_bins=12)
    ax_main.plot(mean_pred, frac_pos, 's-',
                 color=P['cyan'], lw=2, ms=6, label='After calibration', zorder=4)

    # Threshold line
    ax_main.axvline(tg.threshold, color=P['gold'], lw=1.5, ls=':',
                    label=f"Threshold ({tg.threshold:.2f})")

    e_before = tg._pre_report['ECE']
    e_after  = tg._post_report['ECE']
    ax_main.set_title(
        f"{title}  ·  ECE: {e_before:.4f} → {e_after:.4f}  (↓{e_before-e_after:.4f})",
        color=P['white'], pad=8)
    ax_main.set_ylabel('Fraction of Positives')
    ax_main.set_xlim(0,1); ax_main.set_ylim(0,1)
    ax_main.legend(loc='upper left')
    ax_main.grid(True)

    # Histogram
    ax_hist.hist(tg._cc, bins=20, color=P['cyan2'], alpha=0.5, label='Raw')
    ax_hist.hist(cal_confs, bins=20, color=P['green'], alpha=0.5, label='Calibrated')
    ax_hist.axvline(tg.threshold, color=P['gold'], lw=1.5, ls=':')
    ax_hist.set_xlabel('Confidence'); ax_hist.set_ylabel('Count')
    ax_hist.legend(fontsize=8)

    if standalone:
        plt.tight_layout()
        plt.show()


def plot_multi_model_dashboard(models_dict):
    """
    models_dict: {model_name: {'tg': TruthGuard, 'scenario': str}}
    Full side-by-side comparison dashboard.
    """
    n = len(models_dict)
    fig = plt.figure(figsize=(6*n, 14), facecolor=P['bg'])
    fig.suptitle('AEGIS-∞  ·  Multi-Model Enterprise Dashboard',
                 color=P['cyan'], fontsize=15, fontweight='bold', y=0.98)

    outer = gridspec.GridSpec(4, n, figure=fig,
                              hspace=0.45, wspace=0.35)

    colors = [P['cyan'], P['green'], P['gold'], P['purple'], P['orange']]

    for col, (mname, mdata) in enumerate(models_dict.items()):
        tg    = mdata['tg']
        color = colors[col % len(colors)]
        pre   = tg._pre_report
        post  = tg._post_report

        # ── Row 0: Reliability Diagram ─────────────────────────
        inner = gridspec.GridSpecFromSubplotSpec(
            2, 1, subplot_spec=outer[0, col], height_ratios=[3,1], hspace=0.05)
        ax_r = fig.add_subplot(inner[0])
        ax_h = fig.add_subplot(inner[1])
        plot_reliability_diagram(tg, mname, ax_r, ax_h)

        # ── Row 1: Metric Bars ─────────────────────────────────
        ax_m = fig.add_subplot(outer[1, col])
        ax_m.set_facecolor(P['bg3'])
        metrics = ['ECE','MCE','Brier','LogLoss']
        x = np.arange(len(metrics))
        w = 0.35
        bars_pre  = ax_m.bar(x - w/2, [pre[k] for k in metrics],
                             w, color=P['red'], alpha=0.7, label='Before')
        bars_post = ax_m.bar(x + w/2, [post[k] for k in metrics],
                             w, color=color, alpha=0.85, label='After')
        ax_m.set_xticks(x); ax_m.set_xticklabels(metrics, fontsize=8)
        ax_m.set_title(f'{mname} · Metrics', color=P['white'])
        ax_m.legend(fontsize=8); ax_m.grid(True, axis='y')

        # ── Row 2: Confidence Distribution ────────────────────
        ax_d = fig.add_subplot(outer[2, col])
        ax_d.set_facecolor(P['bg3'])
        pos = tg._cc[tg._lc == 1]
        neg = tg._cc[tg._lc == 0]
        bins = np.linspace(0,1,25)
        ax_d.hist(pos, bins, alpha=0.65, color=P['green'], label='Correct')
        ax_d.hist(neg, bins, alpha=0.65, color=P['red'],   label='Wrong')
        ax_d.axvline(tg.threshold, color=P['gold'], lw=2, ls='--',
                     label=f"Thresh={tg.threshold:.2f}")
        ax_d.set_title(f'{mname} · Distribution', color=P['white'])
        ax_d.legend(fontsize=8); ax_d.grid(True, axis='y')

        # ── Row 3: Score Card ──────────────────────────────────
        ax_s = fig.add_subplot(outer[3, col])
        ax_s.set_facecolor(P['bg3'])
        ax_s.axis('off')
        improvement = tg.summary()['improvement']
        lines = [
            ('AUC-ROC',   f"{post['AUC-ROC']:.4f}",  P['cyan']),
            ('ECE ↓',     f"{improvement['ECE_delta']:+.5f}", P['green'] if improvement['ECE_delta'] > 0 else P['red']),
            ('Brier ↓',   f"{improvement['Brier_delta']:+.5f}", P['green'] if improvement['Brier_delta'] > 0 else P['red']),
            ('Overconf',  f"{post['Overconf@80%']:.2%}", P['yellow']),
            ('Conf Gap',  f"{post['ConfGap']:+.4f}",  P['gold']),
            ('Method',    tg.method, color),
            ('Domain',    tg.domain.upper(), color),
            ('Threshold', f"{tg.threshold:.2f}",  P['gold']),
        ]
        for i, (lbl, val, col_) in enumerate(lines):
            ax_s.text(0.05, 0.92 - i*0.115, lbl, transform=ax_s.transAxes,
                      fontsize=9, color=P['grey'], va='top')
            ax_s.text(0.60, 0.92 - i*0.115, val, transform=ax_s.transAxes,
                      fontsize=9, color=col_, va='top', fontweight='bold')
        ax_s.set_title(f'{mname} · Score Card', color=P['white'])

    plt.savefig('/tmp/aegis_dashboard.png', dpi=150, bbox_inches='tight',
                facecolor=P['bg'])
    plt.show()
    banner('✓ Dashboard saved', '/tmp/aegis_dashboard.png', P['green'])


banner('✓ Visualization Engine Loaded', '8 plot types · publication-quality')
```

```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  CELL 4 — PRODUCTION MONITORING ENGINE                              ║
# ╚══════════════════════════════════════════════════════════════════════╝
# Simulates a live deployment: ingests streaming batches,
# tracks drift, fires smart alerts.

class ProductionMonitor:
    """
    Continuous monitoring of deployed AI model confidence.
    Detects: confidence drift · distribution shift · threshold violations.
    """

    def __init__(self, tg: TruthGuard, window=50, drift_alpha=0.05):
        self.tg           = tg
        self.window       = window
        self.drift_alpha  = drift_alpha
        self._batches     = []   # list of dicts
        self._alerts      = []
        self._ref_confs   = tg._cc.copy()   # reference distribution

    def ingest_batch(self, confs_raw, labels=None, batch_id=None):
        """Process one batch of model outputs."""
        confs_raw = np.asarray(confs_raw, float)
        confs_cal = self.tg._cal.transform(confs_raw)
        abstain_rate = float((confs_cal < self.tg.threshold).mean())
        overconf     = float((confs_raw > 0.9).mean())

        # KS drift test vs reference
        ks_stat, ks_p = ks_2samp(self._ref_confs, confs_raw)
        drift_detected = bool(ks_p < self.drift_alpha)

        # Wasserstein distance from reference
        wass = wasserstein_distance(self._ref_confs, confs_raw)

        batch = {
            'batch_id':      batch_id or len(self._batches),
            'n':             len(confs_raw),
            'mean_conf_raw': round(float(confs_raw.mean()), 4),
            'mean_conf_cal': round(float(confs_cal.mean()), 4),
            'abstain_rate':  round(abstain_rate, 4),
            'overconf_rate': round(overconf, 4),
            'ks_stat':       round(float(ks_stat), 4),
            'ks_p':          round(float(ks_p), 4),
            'drift':         drift_detected,
            'wasserstein':   round(float(wass), 4),
            'ece':           round(ece(confs_raw, labels if labels is not None
                                       else (confs_raw > 0.5).astype(int)), 4),
        }
        self._batches.append(batch)

        # Smart alerting
        alerts_fired = []
        if drift_detected:
            alerts_fired.append(('DRIFT',     f"KS p={ks_p:.4f} — distribution shift detected"))
        if abstain_rate > 0.3:
            alerts_fired.append(('ABSTAIN',   f"{abstain_rate:.1%} of batch below threshold"))
        if overconf > 0.6:
            alerts_fired.append(('OVERCONF',  f"{overconf:.1%} of batch > 90% raw confidence"))
        if wass > 0.15:
            alerts_fired.append(('SHIFT',     f"Wasserstein={wass:.4f} — significant shift"))
        for atype, msg in alerts_fired:
            self._alerts.append({'batch': batch['batch_id'], 'type': atype, 'msg': msg})

        return batch, alerts_fired

    def dataframe(self):
        return pd.DataFrame(self._batches)

    def alerts_df(self):
        return pd.DataFrame(self._alerts) if self._alerts else pd.DataFrame(
            columns=['batch','type','msg'])

    def plot_timeline(self):
        df = self.dataframe()
        if df.empty:
            print('No batches yet.'); return

        fig, axes = plt.subplots(4, 1, figsize=(13, 11),
                                 facecolor=P['bg'], sharex=True)
        fig.suptitle('AEGIS-∞  ·  Production Monitoring Timeline',
                     color=P['cyan'], fontsize=13, fontweight='bold')

        specs = [
            ('mean_conf_raw',  'Mean Raw Confidence',   P['red'],   False),
            ('mean_conf_cal',  'Mean Calibrated Conf',  P['cyan'],  False),
            ('abstain_rate',   'Abstain Rate',           P['gold'],  True),
            ('ks_stat',        'KS Drift Statistic',     P['purple'],False),
        ]
        for ax, (col, lbl, clr, pct) in zip(axes, specs):
            ax.set_facecolor(P['bg3'])
            ax.plot(df['batch_id'], df[col], color=clr, lw=2, marker='o', ms=4)
            ax.fill_between(df['batch_id'], 0, df[col], color=clr, alpha=0.12)

            if col == 'mean_conf_raw':
                ax.axhline(self.tg.threshold, color=P['gold'], ls='--', lw=1.2,
                           label=f'Threshold {self.tg.threshold}')
                ax.legend(fontsize=8)
            if col == 'ks_stat':
                ax.axhline(0.1, color=P['red'], ls=':', lw=1, label='Drift Threshold')
                ax.legend(fontsize=8)

            # Mark drift batches
            for _, row in df[df['drift']].iterrows():
                ax.axvspan(row['batch_id']-0.4, row['batch_id']+0.4,
                           color=P['red'], alpha=0.15)

            ax.set_ylabel(lbl, fontsize=9)
            if pct: ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
            ax.grid(True)

        axes[-1].set_xlabel('Batch ID')
        plt.tight_layout()
        plt.show()

banner('✓ Production Monitor Loaded',
       'KS drift · Wasserstein · Abstain rate · Smart alerts')
```

```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  CELL 7 — LIVE PRODUCTION SIMULATION                                ║
# ╚══════════════════════════════════════════════════════════════════════╝
# Simulates 20 production batches with injected drift at batch 12.

banner('⬡ Production Monitoring Simulation',
       '20 batches · drift injected at batch 12 · smart alerts active')

# Fit base model
base_confs, base_labels, _ = ScenarioFactory.generate('well_calibrated', 500, seed=42)
tg_prod = TruthGuard(method='beta', domain='financial')
tg_prod.fit(base_confs, base_labels)

monitor = ProductionMonitor(tg_prod, window=50)
rng = np.random.RandomState(999)
all_alerts = []

for batch_id in range(20):
    # Inject drift after batch 12
    if batch_id < 12:
        batch_confs = np.clip(rng.beta(6, 2, 80), 0.01, 0.99)
    else:
        # Overconfident drift
        batch_confs = np.clip(rng.beta(9, 1, 80) + rng.normal(0.05, 0.03, 80), 0.01, 0.99)

    batch_labels = (rng.uniform(0,1,80) < batch_confs - 0.20).astype(int)
    batch_data, alerts = monitor.ingest_batch(batch_confs, batch_labels, batch_id)
    all_alerts.extend(alerts)

# Print alert summary
if all_alerts:
    alert_html = '<div style="background:#0d1321;border:1px solid #ff4d6d33;border-radius:8px;padding:12px 16px;font-family:monospace;">'
    alert_html += '<div style="color:#ff4d6d;font-weight:bold;margin-bottom:8px;">🚨 SMART ALERTS FIRED</div>'
    for a in all_alerts[:10]:
        badge_colors = {'DRIFT':P['purple'],'ABSTAIN':P['gold'],'OVERCONF':P['red'],'SHIFT':P['orange']}
        col = badge_colors.get(a[0], P['grey'])
        alert_html += f'<div style="margin:4px 0;"><span style="background:{col};color:#080c14;padding:2px 8px;border-radius:4px;font-size:10px;font-weight:bold;">{a[0]}</span> <span style="color:#e8f4f8;font-size:11px;"> {a[1]}</span></div>'
    alert_html += '</div>'
    display(HTML(alert_html))

monitor.plot_timeline()

# Show monitoring dataframe
df_mon = monitor.dataframe()
print(f"\n📊 Monitoring Summary ({len(df_mon)} batches):")
display(df_mon[['batch_id','mean_conf_raw','mean_conf_cal',
                'abstain_rate','ks_stat','drift','wasserstein']].round(4))
```

```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  CELL 8 — DOMAIN RISK HEATMAP  (all domains × all methods)          ║
# ╚══════════════════════════════════════════════════════════════════════╝

domains  = ['general','financial','legal','medical','nuclear']
methods  = ['beta','isotonic','temperature']
scenarios_to_test = ['well_calibrated','overconfident','domain_shift']

fig, axes = plt.subplots(1, 3, figsize=(16, 5), facecolor=P['bg'])
fig.suptitle('AEGIS-∞  ·  ECE Improvement Heatmap: Method × Domain × Scenario',
             color=P['cyan'], fontsize=12, fontweight='bold')

for ax, scen in zip(axes, scenarios_to_test):
    confs, labels, desc = ScenarioFactory.generate(scen, 400, seed=2026)
    matrix = np.zeros((len(methods), len(domains)))
    for i, method in enumerate(methods):
        for j, domain in enumerate(domains):
            try:
                tg = TruthGuard(method=method, domain=domain)
                tg.fit(confs, labels)
                imp = tg.summary()['improvement']['ECE_delta']
                matrix[i, j] = imp
            except Exception:
                matrix[i, j] = 0

    im = ax.imshow(matrix, cmap=AEGIS_CMAP, aspect='auto',
                   vmin=-0.02, vmax=0.15)
    ax.set_xticks(range(len(domains)))
    ax.set_xticklabels([d.upper() for d in domains], fontsize=8)
    ax.set_yticks(range(len(methods)))
    ax.set_yticklabels([m.capitalize() for m in methods], fontsize=8)
    for ii in range(len(methods)):
        for jj in range(len(domains)):
            v = matrix[ii, jj]
            ax.text(jj, ii, f"{v:+.3f}", ha='center', va='center',
                    fontsize=8, fontweight='bold',
                    color=P['bg'] if v > 0.05 else P['white'])
    ax.set_title(f'{scen.upper()}', color=P['white'], fontsize=10)
    fig.colorbar(im, ax=ax, shrink=0.8, label='ECE Improvement ↑')

plt.tight_layout()
plt.show()
banner('✓ Heatmap complete', 'Green = ECE improved after calibration')
```

```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  CELL 9 — AUDIT TRAIL EXPORT                                        ║
# ╚══════════════════════════════════════════════════════════════════════╝
# Exports full audit trail as JSON + CSV for compliance / logging.

# Run a batch audit for demonstration
confs_batch, labels_batch, _ = ScenarioFactory.generate('overconfident', 300, seed=77)
tg_audit = TruthGuard(method='beta', domain='medical')
tg_audit.fit(confs_batch, labels_batch)

# Audit 20 random queries
rng = np.random.RandomState(42)
test_confs = rng.uniform(0.5, 0.99, 20)
contexts   = [f'Query #{i+1}: Patient risk assessment' for i in range(20)]
results    = tg_audit.predict(test_confs, context='Batch audit')
# Note: predict returns a list when given array
for c, r in zip(test_confs, results if isinstance(results, list) else [results]):
    tg_audit.predict(float(c), context=f'Patient risk assessment')

trail = tg_audit.audit_trail()

# Export
trail.to_csv('/tmp/aegis_audit_trail.csv', index=False)
with open('/tmp/aegis_audit_trail.json', 'w') as f:
    json.dump(trail.to_dict(orient='records'), f, indent=2, default=str)

# Display
banner('⬡ AUDIT TRAIL EXPORT', f'{len(trail)} records · CSV + JSON')
display(trail[['audit_id','conf_raw','conf_calibrated','threshold',
               'verdict','risk_level','margin','timestamp']].head(15))

# Summary stats
if not trail.empty:
    verdict_counts = trail['verdict'].value_counts()
    risk_counts    = trail['risk_level'].value_counts()
    display(HTML(f'''
    <div style="background:#0d1321;border:1px solid #1e3a5f;border-radius:8px;
                padding:14px 20px;font-family:monospace;font-size:11px;margin:8px 0;
                display:grid;grid-template-columns:1fr 1fr;gap:16px;">
      <div>
        <div style="color:#ffd700;margin-bottom:6px;">VERDICT DISTRIBUTION</div>
        {''.join(f"<div style='color:#e8f4f8;'>{v}: <span style='color:{P['green'] if v=='ANSWER' else P['red']};font-weight:bold;'>{c}</span></div>" for v,c in verdict_counts.items())}
      </div>
      <div>
        <div style="color:#ffd700;margin-bottom:6px;">RISK DISTRIBUTION</div>
        {''.join(f"<div style='color:#e8f4f8;'>{r}: <span style='color:#00e5ff;font-weight:bold;'>{c}</span></div>" for r,c in risk_counts.items())}
      </div>
    </div>'''))

banner('✓ Exports ready', '/tmp/aegis_audit_trail.csv · /tmp/aegis_audit_trail.json', P['green'])
```
